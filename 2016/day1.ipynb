```python
with open('./day1.txt') as f:
    data = f.read().strip()
```

```python
import numpy as np
```

```python
directions = [
    [ 0, 1], #north
    [ 1, 0], #east
    [ 0,-1], #south
    [-1, 0], #west
]
```

```python
current_dir = 0
directions = np.array(directions)
loc = np.array([0, 0])
for line in data.split(','):
    line = line.strip()
    turn = line[0]
    dist = line[1:]

    if turn == 'R':
        current_dir += 1
    elif turn == 'L':
        current_dir -= 1

    if current_dir < 0:
        current_dir += 4
    elif current_dir > 3:
        current_dir -= 4
    loc += directions[current_dir] * int(dist)
    print(np.sum(np.abs(loc)))
```

```python
current_dir = 0
directions = np.array(directions)
loc = np.array([0, 0])
visited = set()
for line in data.split(','):
    line = line.strip()
    turn = line[0]
    dist = line[1:]

    if turn == 'R':
        current_dir += 1
    elif turn == 'L':
        current_dir -= 1

    if current_dir < 0:
        current_dir += 4
    elif current_dir > 3:
        current_dir -= 4
    for i in range(int(dist)):
        loc += directions[current_dir]
        if str(loc) in visited:
            print(loc,"----")
            print(np.sum(np.abs(loc)))
            break
        visited.add(str(loc))
```
