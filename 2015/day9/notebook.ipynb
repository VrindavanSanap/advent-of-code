We have to find the shortest path that such that we cover each city

Basically a traveling sales man problem

```python
def parse_line(line):
	line_elms = line.split()
	source_city = line_elms[0]
	dest_city = line_elms[2]
	distance = line_elms[4]
	# print(source_city, dest_city, distance)
	return source_city, dest_city, distance
```

```python
cities = set() 
distances = {}
for line in lines:
	source_city, dest_city, distance = parse_line(line)
	cities.add(source_city)
	cities.add(dest_city)
	key = frozenset({source_city, dest_city})
	distances[key] = int(distance)
```

```python
def total_route_distance(route):
	n_cities = len(route)
	total_dist = 0
	for i in range(n_cities -1):
		source_city = route[i]
		dest_city = route[i+ 1]
		dist = distances[frozenset({source_city, dest_city})]
		total_dist += dist
	return total_dist
```

```python
import itertools
all_routes = itertools.permutations(cities)
least_dist = float('inf')
shortest_route = ()
for route in all_routes:
	route_dist = total_route_distance(route)
	if (route_dist < least_dist):
		least_dist = route_dist
		shortest_route = route
print(least_dist, shortest_route)
```

```python
all_routes = itertools.permutations(cities)
greatest_dist = float('-inf') # Initialize to negative infinity
longest_route = ()

for route in all_routes:
    route_dist = total_route_distance(route)
    if (route_dist > greatest_dist): 
        greatest_dist = route_dist
        longest_route = route

print(greatest_dist, longest_route)
```
