```python
with open("day6.txt") as f:
    lines = f.read().strip().split("\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lights = np.zeros((1000,1000))
```

```python
def parse_line(line):
    global lights
    data_points = (line.split())
    instruction = data_points[0]
    print(data_points)
    if (instruction == "turn"):
        x1, y1 = [int(i) for i in data_points[2].split(',')]
        x2, y2 = [int(i) for i in data_points[4].split(',')]
        if data_points[1] == "on":
            lights[x1:x2+1, y1:y2+1] = 1
        elif data_points[1] == "off": 
            lights[x1:x2+1, y1:y2+1] = 0
    if (instruction == "toggle"):
        x1, y1 = [int(i) for i in data_points[1].split(',')]
        x2, y2 = [int(i) for i in data_points[3].split(',')]
        lights[x1:x2+1, y1:y2+1] = np.logical_not(lights[x1:x2+1, y1:y2+1])
    print(x1, y1, x2, y2)
```

```python
for line in lines:
    parse_line(line)
```

```python
plt.imshow(lights)
```
